```python
for i in final_NA.keys():
    final_NA[i]=sum(final_NA[i])
for i in final_NA.keys():
    final_EU[i]=sum(final_EU[i])
for i in final_NA.keys():
    final_JP[i]=sum(final_JP[i])
for i in final_NA.keys():
    final_Other[i]=sum(final_Other[i])
for i in final_NA.keys():
    final_Global[i]=sum(final_Global[i])
```
